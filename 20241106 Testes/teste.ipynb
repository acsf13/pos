```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
df = pd.read_csv('gym_members_exercise_tracking.csv')
df.head(10)
```

```python
base_bpm = df[['Age','Avg_BPM']].to_numpy()
```

```python
base_bpm
```

```python
grafico = px.scatter(base_bpm)
grafico
```

```python
grafico1 = px.scatter(x=base_bpm[:,0], y=base_bpm[:,1])
grafico1
```

```python
scatter_bpm = StandardScaler()
base_bpm = scatter_bpm.fit_transform(base_bpm)
```

```python
base_bpm
```

```python
kmeans_bpm = KMeans(n_clusters=3)
kmeans_bpm.fit(base_bpm)
```

```python
kmeans_bpm
```

```python
centroides = kmeans_bpm.cluster_centers_
centroides
```

```python
scatter_bpm.inverse_transform(base_bpm)
```

```python
scatter_bpm.inverse_transform(centroides)
```

```python
rotulo = kmeans_bpm.labels_
rotulo
```

```python
grafico1 = px.scatter(x=base_bpm[:,0], y=base_bpm[:,1], color=rotulo)
grafico2 = px.scatter(x=centroides[:,0], y=centroides[:,1], size=[15,15,15])
grafico3 = go.Figure(data = grafico1.data + grafico2.data)
grafico3
```
