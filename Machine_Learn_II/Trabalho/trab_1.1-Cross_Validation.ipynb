```python
import pandas as pd
from sklearn.model_selection import cross_val_score, KFold
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the CSV into a DataFrame
df_gym = pd.read_csv("gym.csv")
```

```python
# Display the first few rows of the data
df_gym.head()
```

```python
from sklearn.preprocessing import LabelEncoder

# Initialize the LabelEncoder
le = LabelEncoder()

# Apply Label Encoding to the Gender column
df_gym['Gender_Encoded'] = le.fit_transform(df_gym['Gender'])

# Apply Label Encoding to the Workout_Type column
df_gym['Workout_Type_Encoded'] = le.fit_transform(df_gym['Workout_Type'])

# Show the updated dataframe
df_gym[['Gender', 'Gender_Encoded', 'Workout_Type', 'Workout_Type_Encoded']].head()
```

```python
df_gym_new= df_gym.drop(columns=['Gender', 'Workout_Type'])
df_gym_new.head()
```

```python
#Separe training dataset from the target attribute
X = df_gym_new.drop(columns=['Experience_Level'])  # Variáveis independentes
y = df_gym_new['Experience_Level']  # Variável dependente (alvo)
```

```python
from sklearn.preprocessing import StandardScaler
# Padronizar os atributos preditores
# O K-NN é sensível à escala dos dados, então precisamos normalizar
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled
```

```python
# Configurar o modelo K-NN
knn_model = KNeighborsClassifier(n_neighbors=3)  # Começamos com k=3
```

```python
# Configurar a validação cruzada (10 KFolds)
kf = KFold(n_splits=10, shuffle=True, random_state=42)
```

```python
# Aplicar validação cruzada
cv_scores = cross_val_score(knn_model, X_scaled, y, cv=kf, scoring='accuracy')
# cv_scores = cross_val_score(knn_model, X_scaled, y, cv=kf, scoring='accuracy')
print(cv_scores)
```

```python
# Calcular e exibir os resultados
mean_accuracy = np.mean(cv_scores)
print("Acurácias nos 10 KFolds:", cv_scores)
print("Acurácia média:", mean_accuracy)
```
